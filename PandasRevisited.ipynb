```python
# Long Way
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
#Convert list to a series
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
long_movies = movies[is_long]
```

```python
# OR just use the FILTER method
```

```python
movies[movies.duration>=200]
```

```python
#Just select the genre of long movies
moviesGenre = movies[movies.duration>=200]['genre']
```

# Remove duplicate rows in pandas

```python
users.head()
```

```python
users.shape
```

```python
users.zip_code.duplicated().sum()
```

```python
users.duplicated().sum()  # entire dataframe
```

```python
users.loc[users.duplicated(keep='first'),:]  # Duplicated rows ( only first occurence)
```

```python
users.loc[users.duplicated(keep=False),:] 
```
